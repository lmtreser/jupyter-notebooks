# Funciones matemáticas y sus aplicaciones

Una **función** es una relación entre dos conjuntos, donde a cada valor de entrada (dominio) le corresponde **exactamente un valor de salida** (rango). En ingeniería y ciencias, las funciones describen fenómenos físicos, relaciones matemáticas y comportamientos de sistemas.

## Funciones y sus ecuaciones

1. **Lineales**  
   - **Ecuación:** $f(x) = mx + b$  
   - **Aplicación:** Análisis de circuitos eléctricos de corriente continua.

2. **Cuadráticas**  
   - **Ecuación:** $f(x) = ax^2 + bx + c$  
   - **Aplicación:** Trayectoria de proyectiles, análisis de componentes parabólicos en óptica o antenas parabólicas.

3. **Exponenciales**  
   - **Ecuación:** $f(x) = a \cdot b^x$  
   - **Aplicación:** Análisis de carga/descarga de capacitores, comportamiento de señales eléctricas.

4. **Trigonométricas**  
   - **Ecuaciones:** $f(x) = \sin(x), \cos(x), \tan(x)$  
   - **Aplicación:** Modelado de ondas periódicas (señales AC), análisis de vibraciones, y diseño de sistemas de antenas.

### Ejemplo: Función Lineal
La función es: $$ f(x) = 2x + 3$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función lineal
def funcion_lineal(x):
    return 2 * x + 3

# Rango de valores para x
x = np.linspace(-10, 10, 100)
y = funcion_lineal(x)

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = 2x + 3$", color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Función Lineal", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

## Funciones Cuadráticas

Una función cuadrática tiene la forma general: $$ f(x) = ax^2 + bx + c $$ donde $a$, $b$, y $c$ son constantes.

### **Comportamiento de las Funciones Cuadráticas:**

Las funciones cuadráticas pueden **crecer** y **decrecer** dependiendo del valor de $x$ y de la forma de la parábola que representan.

1. **Forma de la Parábola:**
   - **$a > 0$**: La parábola **abre hacia arriba**. Esto significa que:
     - Hay un **mínimo** (punto más bajo).
     - Antes del vértice (mínimo), la función **decrece**.
     - Después del vértice, la función **crece**.

   - **$a < 0$**: La parábola **abre hacia abajo**. Esto significa que:
     - Hay un **máximo** (punto más alto).
     - Antes del vértice (máximo), la función **crece**.
     - Después del vértice, la función **decrece**.

### **Ejemplos:**

1. **Función Cuadrática que Crece y Decrece:**
   - **$f(x) = x^2$ (donde $a = 1$)**:
     - Mínimo en $x = 0$.
     - Decrece para $x < 0$ y crece para $x > 0$.

2. **Función Cuadrática que Decrece y Crece:**
   - **$f(x) = -x^2 + 4$ (donde $a = -1$)**:
     - Máximo en $x = 0$.
     - Crece para $x < 0$ y decrece para $x > 0$.

Para analizar el comportamiento de una función cuadrática específica, se puede encontrar el vértice utilizando la fórmula $x = -\frac{b}{2a}$ para determinar dónde ocurre el crecimiento o el decrecimiento.

### Ejemplo: Función Cuadrática

La función es: $$ f(x) = -x^2 + 4x + 5 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función cuadrática
def funcion_cuadratica(x):
    return -x**2 + 4 * x + 5

# Rango de valores para x
x = np.linspace(-2, 6, 100)
y = funcion_cuadratica(x)

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = -x^2 + 4x + 5$", color='red')
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Función Cuadrática", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

## Funciones Logarítmicas

Una **función logarítmica** es la **inversa** de la función exponencial. La forma general es: $$ f(x) = \log_b(x) $$ donde $b$ es la **base** del logaritmo y $x > 0$. Significa "¿A qué potencia debo elevar $b$ para obtener $x$?".

### Tipos de Logaritmos

- **Logaritmo decimal**: $ \log_{10}(x) $ (base 10).
  Aplicación: Escala de pH o intensidad sonora en decibelios.

- **Logaritmo natural**: $ \ln(x) = \log_e(x) $ (base $ e $, donde $ e \approx 2.718 $).  
  Aplicación: Modelar procesos de crecimiento continuo, como la desintegración radiactiva o finanzas.

### Propiedades Importantes

1. **Producto**:  
   $\log_b(A \cdot B) = \log_b(A) + \log_b(B)$

2. **Cociente**:  
   $\log_b\left(\frac{A}{B}\right) = \log_b(A) - \log_b(B)$

3. **Potencia**:  
   $\log_b(A^n) = n \cdot \log_b(A)$

4. **Cambio de base**:  
   $\log_b(x) = \frac{\log_k(x)}{\log_k(b)}$, donde $k$ es cualquier base.

### Aplicación de las Funciones Logarítmicas

1. **Escalas Logarítmicas.** En la **escala Richter**, los terremotos se miden logarítmicamente para representar grandes cambios en energía.

2. **Finanzas.** En el cálculo de **interés compuesto continuo**, se usa la función $ A = P \cdot e^{rt} $. Si quieres resolver para $t$, aplicas logaritmos: $ t = \frac{\ln(A / P)}{r} $. Esto indica el tiempo necesario para alcanzar un valor específico con interés continuo.

3. **Ciencias Naturales.** En la descomposición radiactiva, el tiempo necesario para que una sustancia se reduzca a la mitad (vida media) se modela con logaritmos naturales.

### Ejemplo: Función Logarítmica

La función es: $$ f(x) = \log_2(x) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función logarítmica
def funcion_logaritmica(x):
    return np.log2(x)  # Logaritmo en base 2

# Rango de valores para x (debe ser positivo)
x = np.linspace(0.1, 10, 500)  # Evitar log(0) con valores pequeños (>0)
y = funcion_logaritmica(x)

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = \log_2(x)$", color='green')
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Función Logarítmica", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

## Funciones Exponenciales

Una **función exponencial** tiene la forma: $$ f(x) = a \cdot b^x $$ donde:  

- $a$ es el coeficiente inicial (amplitud).
- $b > 0$ es la base.
- $x$ es la variable independiente.

### Propiedades:

1. **Crecimiento** si $b > 1$.  
2. **Decrecimiento** si $0 < b < 1$.  
3. La función nunca cruza el eje $x$ (asíntota horizontal en $y = 0$).  

### **Ejemplo: Función Exponencial**
La función es: $$ f(x) = 2 \cdot 3^x $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función exponencial
def funcion_exponencial(x):
    return 2 * (3 ** x)  # Coeficiente a = 2, base b = 3

# Rango de valores para x
x = np.linspace(-2, 2, 500)  # Valores de -2 a 2
y = funcion_exponencial(x)

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = 2 \cdot 3^x$", color='purple')
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")
plt.title("Función Exponencial", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

### Relación entre Exponenciales y Logarítmicas

- **Inversas:**  
  La función logarítmica es la inversa de la función exponencial. Esto significa que si tienes $ y = b^x $, puedes encontrar $ x $ usando logaritmos: $ x = \log_b(y) $
